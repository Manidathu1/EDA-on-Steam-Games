# Pie chart

```python
# pie chart to know the how much percentage of games support and not support english language
english_dict = dict(steam.english.value_counts())
labels = 'English', 'Non-english'
sizes = [english_dict[1], english_dict[0]]
colors = ['gold', 'lightskyblue']
explode = (0, 0.2)  # explode 1st slice

# Plot
with sns.color_palette("colorblind"):
    plt.pie(sizes, explode=explode, labels=labels, # colors=colors,
    autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')
plt.show()
```

```python
# pie chart to for required_age column
age_dict = dict(steam.required_age.value_counts())
labels = 'no age limit', 'Have an age limit'
sizes = [age_dict[0], age_dict[18] + age_dict[16] + age_dict[12] + age_dict[7] + age_dict[3]]
explode = (0, 0.15)

with sns.color_palette("colorblind"):
    plt.pie(sizes, explode=explode, labels=labels,
    autopct='%1.1f%%', shadow=True, startangle=140)


plt.axis('equal')
plt.show()
```

most of the games can be played by any age person but 2.2% of games in the dataset have age limit to play 

```python
# pie chart to show paid and free games in price column of every game
free_dict = dict(steam.price.value_counts())
labels = 'Free', 'Paid'
sizes = [free_dict[True], free_dict[False]]
explode = (0.2, 0)
with sns.color_palette('colorblind'):
    plt.pie(sizes, explode=explode, labels=labels,
    autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis('equal')
plt.title('Piechart of free/paid games among all games')
plt.show()
```
